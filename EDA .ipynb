## Start Time

```python
df.Start_Time[0]
```

```python
#converting date which was in object format to date time format
df.Start_Time = pd.to_datetime(df.Start_Time)
df.Start_Time
```

```python
#extracting hour variable from date time
df.Start_Time.dt.hour
```

### Accident distribution based on the time of the day

### Q) What time of the day are accidents most frequent?


```python
plt.figure(figsize = (10,6))
plt.title('Distribution of Traffic throughout the day')
sns.distplot(df.Start_Time.dt.hour, bins = 24);
```

#### Most accidents happen around 3pm to 6 pm and next from 7 am to 9 am.


```python
# finding max load of traffic wrt day of week

day_of_week = df.Start_Time.dt.dayofweek
day_of_week
```

### Accident distribution throughout the week

### Q) Which days of the week have the most accidents?

```python
plt.title('Distribution of Traffic throughout the week')
sns.distplot(day_of_week, bins = 7)
```

#### Weekends have lower accidents.


### Q) Is the distribution of accidents throughout the day same for weekends and weekdays?

```python
#Is the distribution of accidents throughout the day same for weekends and weekdays?
#get weekend data

sunday_df_Start_Time = df.Start_Time[df.Start_Time.dt.day_of_week == 6]
sunday_df_Start_Time
```

```python
#plotting graph of distribution of traffic on sunday
plt.title('Distribution of traffic on Sunday')
sns.distplot(sunday_df_Start_Time.dt.hour, bins = 24);
```

```python
#monday distribution
monday_df_Start_Time = df.Start_Time[df.Start_Time.dt.day_of_week == 1]
```

```python
#plotting graph of distribution of traffic on monday
plt.title('Distribution of traffic on Monday')
sns.distplot(monday_df_Start_Time.dt.hour, bins = 24, kde = False, norm_hist = True);
```

#### On sundays peak occurs between 1 pm to 6 pm and from 12 am to 3 am.
#### On mondays max traffics happen between 3 pm to 5pm and from 7 am to 9 am.

## Temperature

### Do more accidents happen in warmer or colder areas?
Warm : >= 80 F
Mod : 50-80 F
Low : <= 50 F

```python
#Temperature column
columns = df.columns.values.tolist()
temp_df = df['Temperature(F)']
temp_df
```

```python
#plotting distribution ofaccidents wrt temperature
sns.distplot(temp_df)
```

```python
import matplotlib.pyplot as plt
```

```python
hot_temp = temp_df[temp_df >= 80]
mod_temp = temp_df[temp_df < 50]
```

```python
plt.figure(figsize = (12,6))
plt.title('Distribution of accidents wrt diff temperatures')
plt.plot(hot_temp)
plt.plot(mod_temp)
plt.xlabel('Accidents')
plt.ylabel('Temp(F)')


plt.show()
```
